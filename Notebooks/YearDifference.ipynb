```python
# cargamos los csv del numero de reviews de cada juego por a√±o
df2013 = pd.read_csv("year2013.csv")
df2014 = pd.read_csv("year2014.csv")
df2015 = pd.read_csv("year2015.csv")
df2016 = pd.read_csv("year2016.csv")
df2017 = pd.read_csv("year2017.csv")
df2018 = pd.read_csv("year2018.csv")
df2019 = pd.read_csv("year2019.csv")
df2020 = pd.read_csv("year2020.csv")
df2021 = pd.read_csv("year2021.csv")
years = [df2013,df2014,df2015,df2016,df2017,df2018,df2019,df2020,df2021]
```

```python
reviews_per_year = [i['count'].sum() for i in years]
reviews_per_year
```

```python
# numero de reviews por a√±o (2021 no habia acabado al crearse el dataset)
plt.ticklabel_format(style='plain', axis='y')
plt.plot([2013,2014,2015,2016,2017,2018,2019,2020,2021], reviews_per_year)
plt.savefig('graficos/NumReviewsPerYear.png')
```
